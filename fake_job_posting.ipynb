job_id - Уникальный идентификатор работы.

company_profile - Краткое описание компании.

benefits - Зачислены предлагаемые работодателем льготы.

has_company_logo - Верно, если присутствует логотип компании.

has_question - Истинно, если присутствуют проверочные вопросы.

employment_type - Полный тип, неполный рабочий день, контракт и т.д.

required_experience - Руководитель, Начальный уровень, Стажер и т.д.

required_education - Докторантура, Магистр, Бакалавр и т.д.

industry - Автомобилестроение, IT, Здравоохранение, Недвижимость и т.д.

function - Консалтинг, инжиниринг, исследования, продажи и т. д.

```python
fig, ax = plt.subplots(figsize=(11, 11))

data_of_edc = data['required_education'].value_counts()

data_of_edc

small_edc_totals = data_of_edc[data_of_edc < 50]
big_edc_totals = data_of_edc[data_of_edc >= 50]
small_sums = pd.Series([small_edc_totals.sum()], index=["Other"])
big_edc_totals = big_edc_totals._append(small_sums)
big_edc_totals.plot(kind="pie", label="", autopct='%.1f')
plt.title('Соотношение требуемого образования на круговой диаграмме')
```

```python
graph2 = sns.countplot(y='telecommuting', data=data)
plt.title('Количество вакансий связанных с удалённой работой')
```

```python
df = {
    "required_education": pd.Series(data['required_education']),
    "fraudulent": pd.Series(data['fraudulent']),

}

data_of_edc_fraud = pd.DataFrame(df)
data_of_edc_fraud_1 = data_of_edc_fraud.query('fraudulent == 1')

fig, ax = plt.subplots(figsize=(11, 11))

labels = data_of_edc_fraud_1.value_counts()
labels.plot(kind='pie', label="",  autopct='%.1f')
plt.title("Соотношение требуемого образования на круговой диаграмме для фейковых вакансий")
```

```python
dep = {
    "department": pd.Series(data['department']),
    "fraudulent": pd.Series(data['fraudulent']),
}

data_of_dep_fraud = pd.DataFrame(dep)
data_of_dep_fraud_1 = data_of_dep_fraud.query('fraudulent == 1')


fig, ax = plt.subplots(figsize=(20, 20))

labels = data_of_dep_fraud_1.value_counts()

big_labels = labels[labels > 1]


big_labels.plot(kind='bar', label="")
plt.ylabel("Количество")
plt.title("Колличество фейковых вакансий по отдедам")
small_labels = labels[labels == 1]
small_labels
```

```python
ig, ax = plt.subplots(figsize=(20, 20))
data_emp = data['employment_type'].value_counts()
data_emp.plot(kind="bar", label="")
plt.title("Колличество вакансий по виду занятости")

emp = {
    "employment_type": pd.Series(data['employment_type']),
    "fraudulent": pd.Series(data['fraudulent']),
}
data_of_emp_fraud = pd.DataFrame(emp)
data_of_emp_fraud_1 = data_of_emp_fraud.query('fraudulent == 1')

fig, ax = plt.subplots(figsize=(20, 20))

emp_label = data_of_emp_fraud_1.value_counts()

emp_label.plot(kind='bar', label="")
plt.ylabel("Количество")
plt.title("Колличество фейковых вакансий по виду занятости")
```
